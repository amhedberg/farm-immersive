```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# sheet1: 2019 Summary
df = pd.read_excel('~/Documents/farm-immersive/farm-immersive/2019EBAAdditionalData.xlsx', sheet_name=0, header=1)
```

```python
df.head()
```

```python
df.Region.unique()
```

```python
sub_sahara = df[df['Region'].str.contains("Sub-Sahara")]
# africa = df[df['Region'].str.contains("Africa")]
```

```python
sub_sahara.head()
```

```python
sns.set_context("talk", font_scale=1.0)
plt.figure(figsize=(10,6))
sns.scatterplot(data=sub_sahara, x="Supplying seed (0-24)", y="Accessing finance (0-16)", hue="Income Group")
plt.legend(bbox_to_anchor=(1.01, 1),borderaxespad=0)
plt.title("Sub-Saharan Africa")
plt.tight_layout()
```

```python
sns.set_context("talk", font_scale=1.0)
# plt.figure(figsize=(5,25))
g = sns.catplot(
    data=df, kind="bar",
    x="Supplying seed (0-24)", y="Income Group", hue="Region",
    ci="sd", palette="dark", alpha=.6, height=6
)
g.despine(left=True)
g.set_axis_labels("Supplying seed", "")
g.legend.set_title("")
# plt.legend(bbox_to_anchor=(1.01, 1),borderaxespad=0)
# plt.title("Sub-Saharan Africa")
plt.tight_layout()
```

```python
sns.set_context("talk", font_scale=0.95)
plt.figure(figsize=(8,7))
sns.barplot(data=df, x="Supplying seed (0-24)", y="Income Group", hue="Region")
plt.legend(bbox_to_anchor=(1.01, 0.5),borderaxespad=0)
```

```python
sns.set_context("talk", font_scale=0.95)
plt.figure(figsize=(8,7))
sns.barplot(data=df, x="Supplying seed (0-24)", y="Region", hue="Income Group")
plt.legend(bbox_to_anchor=(1.01, 0.5),borderaxespad=0)
```

```python
sns.set_context("talk", font_scale=1)
# plt.figure(figsize=(5,15))
sns.catplot(data=df, x="Supplying seed (0-24)", y="Region", kind="violin", height=6, aspect=2)
# plt.legend(bbox_to_anchor=(1.01, 0.5),borderaxespad=0)
```
